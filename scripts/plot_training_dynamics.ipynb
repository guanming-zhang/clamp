```python
import csv
from matplotlib import pyplot as plt
```

```python
csv_file = "/home/guanming/HPC-Scratch/sig-ml/clap/simulations/imagenet/linear/logloss_momentum/dir3#DATA#-dataset-IMAGENET1K-n_views-4-#SSL#-lr-0.9-batch_size-1024-momentum-0.9-weight_decay-1e-06-lw0-0.0-lw2-0.0-rs-7.0-run-00/ssl/logs/csv/version_0/metrics.csv"
val = {"val_acc":[],"val_radius":[],"val_num_nbr":[]}
with open(csv_file, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        for k in val:
            value = row.get(k)
            if value:  # skip empty or None
                try:
                    val[k].append(float(value))
                except ValueError:
                    continue  # skip invalid values
```

```python
fig = plt.figure(figsize=(5, 4))
ax1 = fig.add_axes(111)
ax1.plot(val["val_num_nbr"],linewidth=2.0)
ax1.set_xlabel("epoch",fontsize=20)
ax1.set_ylabel("number of neighbours",fontsize=20)
ax1.tick_params(axis='both', labelsize=20)
plt.savefig("val_num.png",bbox_inches='tight',dpi=600) 
```

```python
fig = plt.figure(figsize=(5, 4))
ax2 = fig.add_axes(111)
ax2.plot(val["val_radius"],linewidth=2.0)
ax2.set_xlabel("epoch",fontsize=20)
ax2.set_ylabel("sub-manifold size",fontsize=20)
ax2.tick_params(axis='both', labelsize=20)
plt.savefig("val_radius.png",bbox_inches='tight',dpi=600) 
```
