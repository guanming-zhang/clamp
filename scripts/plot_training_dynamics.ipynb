```python
import csv
from matplotlib import pyplot as plt
```

```python
csv_file = "/home/guanming/HPC-Scratch/sig-ml/clap/simulations/imagenet/linear/logloss_no_momentum/dir0#DATA#-n_views-4-#SSL#-lr-8.0-batch_size-1024-momentum-0.0-weight_decay-1e-06-lw0-0.0-lw2-0.0-rs-7.0-run-00/ssl/logs/csv/version_0/metrics.csv"
val = {"val_acc":[],"val_radius":[],"val_num_nbr":[]}
with open(csv_file, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        for k in val:
            value = row.get(k)
            if value:  # skip empty or None
                try:
                    val[k].append(float(value))
                except ValueError:
                    continue  # skip invalid values
```

```python
fig = plt.figure()
ax1 = fig.add_axes(111)
ax1.plot(val["val_num_nbr"])
```
